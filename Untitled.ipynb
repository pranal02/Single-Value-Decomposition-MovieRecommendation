```python
#imprt the libraries
import pandas as pd
import numpy as np
```

```python
# Reading the rating file
ratings= pd.read_csv('ratings.csv')
ratings.head()
```

```python
movies =pd.read_csv('movies.csv'  )
movies.head()
```

```python
n_users = ratings.userId.unique().shape[0]
n_users
```

```python
n_movies = ratings.movieId.unique().shape[0]
print(f'Number of users ={n_users} and number of  movies={n_movies}')
```

```python
# pivot is used to convert the table as the fetature we want to know as index

Ratings = ratings.pivot(index='userId' ,columns='movieId',values='rating').fillna(0)
Ratings.head()
```

```python
# go to anaconda promt then go to current working directory and  use = install -c conda-forge scikit-surprise
```

```python
from surprise import Reader,Dataset,SVD
from surprise.model_selection import cross_validate # TO CCHECK THE PERFORMANCE OF MODEL

#Load Reader Library
reader = Reader()

#Load rating dataset with Dataset library
data= Dataset.load_from_df(ratings[['userId','movieId','rating']],reader)

#use the SVD algorithm.
svd=SVD()

cross_validate(svd,data,measures=['RMSE','MAE'],cv=3,verbose=True)
```
